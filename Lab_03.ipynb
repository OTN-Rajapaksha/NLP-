```python
# Define positive and negative words
positive_words = {"love", "happy", "great", "amazing", "good", "like", "fantastic", "wonderful"}
negative_words = {"hate", "bad", "terrible", "awful", "horrible", "sad", "worst", "dislike"}

# Function to classify text
def classify_sentiment(text):
    words = text.lower().split()
    pos_score = sum(word in positive_words for word in words)
    neg_score = sum(word in negative_words for word in words)

    if pos_score > neg_score:
        return "Positive"
    elif neg_score > pos_score:
        return "Negative"
    else:
        return "Neutral"

# Test inputs
texts = [
    "I love this movie, it’s amazing!",
    "The place was awful and sad.",
    "Great service but bad food.",
    "I love the food but the service was terrible"
]

# Print outputs
for text in texts:
    print(f"Text: {text}\nLexicon Sentiment: {classify_sentiment(text)}\n")
```

```python
# Install transformers library (only run once)
!pip install -q transformers

# Import and use pretrained model
from transformers import pipeline

# Load sentiment pipeline
classifier = pipeline("sentiment-analysis")

# Test examples
texts = [
    "I absolutely love this product!",
    "The food was horrible and I hated the service.",
    "Great service but bad food."
]

# Print predictions
for text in texts:
    result = classifier(text)[0]
    print(f"Text: {text}\nHugging Face Sentiment: {result['label']} (Score: {result['score']:.2f})\n")
```
